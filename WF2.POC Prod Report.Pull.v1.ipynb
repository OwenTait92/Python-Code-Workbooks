```python
from seeq import spy
import pandas as pd
from seeq.spy.assets import Asset
#To automatically run all lines of code, hit the "restart the kernel, then re-run the whole notebook (with dialog)" button (double right arrow)
```

# 1. Import workbook/path and set time period

```python
workbookName = 'Brintnell Prod1 Report' #Edit Required-> change to same name as Step #1 in WF1.NVA Report.Hierarchy
pathName = 'POC Prod Report'#Edit Required-> change to same name as Step #3 in WF1.NVA Report.Hierarchy for meta_data_df['Build Path']
et = pd.Timestamp.now(tz='US/Mountain') #Edit Required-> This is the End Time, can enter a spefic date if desired
st = et -  pd.Timedelta(days=1) #Edit Required-> This is the Start Time (1 days prior End Time), can enter a spefic date if desired
```

# 2. Searches for Signal/Removes Duplicates

```python
POC_Prod_search = spy.search({
                        'Path' : pathName,
                        'Name' : 'Liquid Flow Rate',#Edit Required-> From WF1.NVA Report.Hierarchy Step #1, enter the signal_search_name 
                        }, workbook = workbookName, quiet = True
                      ).drop_duplicates(subset = ["Asset"],keep = "first")[:]  #removes the duplicates and only keep the first from all [:] rows (can change this to [:3] if you only want the first 3)

POC_Prod_search
```

# 3. Create Custom Asset List to loop through

```python
POC_Prod_list = (POC_Prod_search['Path'] + ' >> ' + POC_Prod_search['Asset'] ).tolist() #creates a list of assets for loop
POC_Prod_list
```

# 4. Loop through custom Asset

```python
dfResult = []
for n in POC_Prod_list:

    searchresult = spy.search({'Path' : n}, workbook = workbookName, recursive = False, quiet = True)
    pullresult = spy.pull(searchresult, start=st, end = et, grid='1h', shape = 'auto', header = 'Name', group_by=['Asset'], quiet = False)
    pullresult = pullresult.assign(Pad = n.split(">>")[2].strip())
    dfResult.append(pullresult)

#loops through each well from the POC_Prod_list in Step #3. References your Start and End times  
    
dfreport = pd.concat(dfResult)
#Creates the report
```

```python
#reset index and name timestamp column
dfreport = dfreport.reset_index().rename(columns={'level_0' : 'timestamp_tz'})
```

```python
#create timestamp column without timezone
dfreport['timestamp'] = dfreport['timestamp_tz'].apply(lambda x: pd.Timestamp(x).tz_localize(None))
```
